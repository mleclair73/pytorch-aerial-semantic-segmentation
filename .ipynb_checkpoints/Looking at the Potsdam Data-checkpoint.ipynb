```python
# How is the data described
"""
Six categories/classes have been defined:

Impervious surfaces (RGB: 255, 255, 255)
Building (RGB: 0, 0, 255)
Low vegetation (RGB: 0, 255, 255)
Tree (RGB: 0, 255, 0)
Car (RGB: 255, 255, 0)
Clutter/background (RGB: 255, 0, 0)
Undefined (RGB: 0, 0, 0)

The clutter/background class includes water bodies (present in two images with part of a river)
and other objects that look very different from everything else(e.g., containers, tennis courts,
swimming pools) and that are usually not of interest in semantic object classification in urban scenes
"""
categorical_classes = {
    (255, 255, 255): 0,
    (  0,   0, 255): 1,
    (  0, 255, 255): 2,
    (  0, 255,   0): 3,
    (255, 255,   0): 4,
    (255,   0,   0): 5,
    (  0,   0,   0): 6,
}

class_names = {
    "impervious": (255, 255, 255),
    "building"  : (  0,   0, 255),
    "low_veg"   : (  0, 255, 255),
    "tree"      : (  0, 255,   0),
    "car"       : (255, 255,   0),
    "background": (255,   0,   0),
    "undefined" : (  0,   0,   0),
}
def to_hex(rgb_tuple):
    return '#' + ''.join([f"{y:02x}" for y in rgb_tuple])

def make_legend(p=None):
    patches = []
    for i, (name, color) in enumerate(class_names.items()):
        patches.append(mpatches.Patch(color=to_hex(color), label=name))
    if p:
        p.legend(handles=patches)
    else:
        plt.legend(handles=patches)
```

```python
# Let's extract categorical labels from the data
def to_categories(labels, channel_first=False):
    # Labels is c, h, w
    shape = labels.shape[1:] if channel_first else labels.shape[:-1]
    reshape = [3, 1, 1] if channel_first else [1,1,3] 
    ax = 0 if channel_first else 2
    categories = np.zeros(shape, dtype=np.uint8)
    for k, v in categorical_classes.items():
        mask = np.all(labels == np.array(k).reshape(*reshape), axis=ax)      
        categories[mask] = v
    return categories

def to_colors(categories):
    color = np.zeros((categories.shape[0], categories.shape[1], 3), dtype=np.uint8)
    for k, v in categorical_classes.items():
        color[categories == v] = k
    return color

cats = to_categories(labels)
plt.imshow(to_colors(cats))
```

```python
def calculate_mean_std(ids, data_dir):
    files = [os.path.join(data_dir, f'top_potsdam_{idx}_RGB.tif')for idx in ids]
    running_mean = []
    running_std = []
    for file in files:
        print("Analyzing", file.split('/')[-1])
        # Scale 0-1 (like ToTensor)
        img = io.imread(file)
        mean, std = cv2.meanStdDev(img)
        running_mean.append(mean)
        running_std.append(std)
    return {'mean': np.array(running_mean).mean(axis=0),
            'std':  np.array(running_std).mean(axis=0)}
# print(calculate_mean_std(train_ids, DATA_DIR))
```

```python
# Next we define a PyTorch Dataset that will allow us to work with the ISPRS Potsdam Data
# This is a subclass of torch.utils.data.Dataset https://pytorch.org/docs/stable/data.html

class PotsdamDataset(torch.utils.data.Dataset):
    def __init__(self, ids, data_dir, label_dir, test=None):
        super(PotsdamDataset, self).__init__()
        
        # List Files
        self.data_files = [os.path.expanduser(os.path.join(data_dir, f'top_potsdam_{idx}_RGB.tif')) 
                           for idx in ids]
        self.label_files = [os.path.expanduser(os.path.join(label_dir, f'top_potsdam_{idx}_label.tif')) 
                            for idx in ids]
        
        # Validate the Input
        invalid_files =[f for f in self.data_files and self.label_files if not os.path.isfile(f)] 
        if invalid_files:
            raise OSError(f'The following files: {invalid_files} are not valid')
        
        # Create a chache for the images to minimize IO
        # This read takes ~500ms, so we can cache. Might blow RAM up to Dataset Size
        self.cache_ = {}
        
        if test:
            self.transform = transforms.Compose([
                transforms.ToPILImage(),
                transforms.CenterCrop((500,500)),
                transforms.ToTensor()
            ])
        
    def __len__(self):
        """This will define the length of an epoch / size of a batch"""
        return len(self.data_files) #10000
    
    def __getitem__(self, i):
        """PyTorch uses this getter to access data through the class. All data augmentation 
        happens here on retreival"""
        
        idx = random.randint(0, len(self.data_files) - 1)
        if idx not in self.cache_.keys():
            
            # Load Sample
            img = io.imread(self.data_files[idx])
            label = io.imread(self.label_files[idx])
            if self.transform:
                img, label = self.transform(img).numpy(), self.transform(label).numpy()
            img = img.astype(np.float64)
            img = img - np.array([86.58021059, 92.42789518, 85.77866355])[:, None, None]
            img = np.copy(img[::-1, :, :])
            label = torch.from_numpy(to_categories(label, channel_first=True)).float()
            img = torch.from_numpy(img).float()
            sample = (img, label)
            
        return sample
```

```python
def get_all_base_layers(module):
    acc = {}
    for name, child in module.named_children():
        base_layers = get_all_base_layers(child)
        if not base_layers:
            acc[name] = child
        else:
            acc = {**acc, **base_layers}
    return acc
```

```python
# This is NOT deterministic, there are batchnorm variables that are un-initialized. Potentially other problems?
# Maybe it's possible to retreive these but do we need to?
# https://discuss.pytorch.org/t/model-eval-gives-incorrect-loss-for-model-with-batchnorm-layers/7561/30
# https://discuss.pytorch.org/t/batchnorm-eval-cause-worst-result/15948
weights_folder = 'models/weights'
weights_files = [os.path.join(weights_folder, weight) for weight in 
                    ['icnet_cityscapes_train_30k.caffemodel.pth',
                     'icnet_cityscapes_train_30k_bnnomerge.caffemodel.pth',
                     'icnet_cityscapes_trainval_90k.caffemodel.pth',
                     'models/weights/icnet_cityscapes_trainval_90k.caffemodel.pth']
                ]

def load_weights(model, weights_path, verbose=False):

    pre_trained = torch.load("models/weights/icnet_cityscapes_trainval_90k.caffemodel.pth")
    model_weights = model.state_dict()
    missing_layers = {'bn':{}, 'bias':{}, 'misc':{}}
    mapped_layers = {'.'.join(k.split('.')[-2:]): k for k in model.state_dict().keys()}

    for src, dst in mapped_layers.items():
        if src in pre_trained:
            model_weights[dst] = pre_trained[src]
        elif 'bn' in src:
            missing_layers['bn'][src] = dst
        elif 'bias' in src:
            missing_layers['bias'][src] = dst
        else:
            missing_layers['misc'][src] = dst
    if verbose:
        print(missing_layers)
    model.load_state_dict(model_weights)
    return model
    
model = load_weights(model, weights_files[0])
```
